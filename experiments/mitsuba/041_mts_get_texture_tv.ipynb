```python
def wandb_init(name):
    run = wandb.init(
        project = "mts_texture_test", 
        reinit = True,
        name = name,
        config={
            "steps": 100,
            "learning_rate": 0.2,
            "optimizer": "Adam",
            "noise_width": 200,
            "noise_height": 400,
        }
    )
    
    return run 
```
